```python
import numpy as np

# Повредим 0.5-3% данных в колонках
min_percentage_to_modify = 0.005
max_percentage_to_modify = 0.03

# Выберем случайное количество строк в заданном диапазоне
rows_to_modify = df.sample(frac=np.random.uniform(min_percentage_to_modify, max_percentage_to_modify)).index

# Для каждой строки выбериаем случайную колонку и сделаем ячейку в этой колонке пустой
for row in rows_to_modify:
    column_to_modify = df.columns[np.random.randint(0, len(df.columns))]
    df.at[row, column_to_modify] = np.nan
```
